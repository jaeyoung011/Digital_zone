```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, Y_train, Y_test =train_test_split(X, Y,
                                                   test_size=0.2,random_state=1234)

smote = SMOTE(random_state=0) # train 만 늘려주기
X_train,Y_train = smote.fit_resample(X_train,Y_train)

print(X_train.shape)
print(X_test.shape)
```

```python
pipe_list = [('scaler',MinMaxScaler()),
              ('model', LGBMClassifier(n_estimators=10))]
pipe_model = Pipeline(pipe_list)
```

```python
grid_parameter = {'model__class_weight':[None,'balanced'],
                 'model__max_depth':range(3,6),
                 'model__min_samples_leaf':range(3,6),
                 'model__min_samples_split':range(3,6)}


grid_model = GridSearchCV(pipe_model, cv=10, param_grid= grid_parameter)
grid_model.fit(X_train, Y_train)
```

```python
best_model = grid_model.best_estimator_
```

```python
best_model
```

```python
Y_train_pred = best_model.predict(X_train)
Y_test_pred  = best_model.predict(X_test)
```

```python
print(classification_report(Y_train , Y_train_pred))
```

```python
print(classification_report(Y_test , Y_test_pred))
```
