```python
import pandas as pd 
import warnings
warnings.filterwarnings(action='ignore') 
```

```python
df = pd.read_csv('../final_data.csv')
df.head(2)
```

```python
df['순번'] =  df['순번'].astype('category')
df['순번'] = df.순번.cat.codes
```

```python
df.sort_values(['순번','semester'], ascending=[True,True], kind='stable', inplace=True)
```

```python
df = df[['순번','단과명','직무중분류','semester','교과목명','이수구분','성적','학점']]
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df_list = df.values.tolist()
```

```python
df_list
```

```python
def concat_horizontally(index):
    ls = []
    for i in df_list:
        if i[0] == index: # and i[3] < 8:
            ls.append(i[3:])
    
    cond = (df['순번'] == index)
    dep = df.loc[cond]['단과명'].tolist()[0]
    exp = df.loc[cond]['직무중분류'].tolist()[0]
    
    data = {'순번': [index], '단과명': [dep], '직무중분류': [exp]}
    df_new = pd.DataFrame(data)
    
    count = 0
    for i in range(len(ls)-1):
        count += 1
        #df_new['semester{}-{}'.format(ls[i][0],i+1)] = ls[i][0]
        df_new['교과목명{}-{}'.format(ls[i][0],count)] = ls[i][1]
        df_new['이수구분{}-{}'.format(ls[i][0],count)] = ls[i][2]
        df_new['성적{}-{}'.format(ls[i][0],count)] = ls[i][3]
        df_new['학점{}-{}'.format(ls[i][0],count)] = ls[i][4]
        if ls[i][0] != ls[i+1][0]:
            count = 0
    
    return df_new
```

```python
df_final = concat_horizontally(0)
for i in range(1,1068):
    df_final = pd.concat([df_final, concat_horizontally(i)],
              ignore_index=True, join='outer')
```

```python
col = df_final.columns[3:]
for i in col:
    if i[-2:] >= '10' or i[-1:] >= '8':
        df_final.drop(columns=[i], inplace=True)
```
